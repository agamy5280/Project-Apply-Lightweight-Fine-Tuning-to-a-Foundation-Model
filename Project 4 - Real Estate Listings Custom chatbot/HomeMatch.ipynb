## 2. Generating Real Estate Listings

```python
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

# Prompt to generate listings
prompt = """
Generate 20 diverse real estate listings. Each listing should include the following details:

- **Neighborhood**: Use a mix of urban, suburban, and rural areas with varying characteristics.
- **Price**: Include a range of prices from affordable starter homes to luxurious estates.
- **Bedrooms**: Vary these from small apartments (e.g., 1-bedroom) to large family homes (e.g., 5-bedroom).
- **Bathrooms**: Include the number of bathrooms to match the property.
- **House Size**: Use a range of sizes in square feet, from compact city apartments to sprawling countryside estates.
- **Description**: Provide a creative and appealing description, highlighting unique features such as large backyards, modern kitchens, hardwood floors, smart home features, or stunning views.
- **Neighborhood Description**: Offer a compelling overview of the neighborhood, mentioning amenities, nearby attractions, lifestyle (e.g., quiet and family-friendly vs. vibrant and bustling), and proximity to key locations (e.g., schools, parks, shopping centers, or downtown areas).

Ensure that each listing is formatted like this with the listings must be in the JSON array of dictionaries with each item in the format as shown below:

[
  {
    "location": Green Oaks,
    "list_price": 500000,  # Integer value
    "bedrooms": 3,  # Integer value
    "bathrooms": 2,  # Integer or float value
    "square_feet": 2000,  # Integer value
    "description": Welcome to this eco-friendly oasis nestled in the heart of Green Oaks. This charming 3-bedroom, 2-bathroom home boasts energy-efficient features such as solar panels and a well-insulated structure. Natural light floods the living spaces, highlighting the beautiful hardwood floors and eco-conscious finishes. The open-concept kitchen and dining area lead to a spacious backyard with a vegetable garden, perfect for the eco-conscious family. Embrace sustainable living without compromising on style in this Green Oaks gem.
    "neighborhood_description": Green Oaks is a close-knit, environmentally-conscious community with access to organic grocery stores, community gardens, and bike paths. Take a stroll through the nearby Green Oaks Park or grab a cup of coffee at the cozy Green Bean Cafe. With easy access to public transportation and bike lanes, commuting is a breeze.
  },
  ...
]

Generate the listings in the format above, using varied buyer preferences (e.g., homes for families, singles, retirees, young professionals), and include properties suited for various lifestyles (e.g., eco-friendly, luxury, waterfront, or close to nightlife).
"""

# Get the response from the model
response = llm.predict(prompt)

try:
    listings = json.loads(response)
    
    # Write the listings to a JSON file
    with open("listings.json", "w") as f:
        json.dump(listings, f, indent=4)
        
    print("Listings saved successfully.")
except json.JSONDecodeError as e:
    print(f"Error in parsing JSON: {e}")
```

## 3. Storing Listings in LanceDB

```python
# Load listings into a Pandas DataFrame
listings_df = pd.read_json("listings.json")

# Initialize embeddings model
embeddings = OpenAIEmbeddings()

# Initialize LanceDB client
db = lancedb.connect("homematch_db")  # Path to the LanceDB database

# Drop the table if it exists
if "listings" in db.table_names():
    db.drop_table("listings")
    print("Dropped the existing 'listings' table.")

# Define the schema for the vector database
class RealEstateListings(LanceModel):
    location: str
    list_price: float
    bedrooms: float
    bathrooms: float
    square_feet: float
    description: str
    neighborhood_description: str
    description_vector: vector(1536)  # Size of the embedding vector

# Function to generate embeddings
def get_embedding(text: str) -> list:
    return embeddings.embed_query(text)

# Prepare data with embeddings
data = listings_df.apply(
    lambda row: {
        "location": row["location"],
        "list_price": row["list_price"],
        "bedrooms": row["bedrooms"],
        "bathrooms": row["bathrooms"],
        "square_feet": row["square_feet"],
        "description": row["description"],
        "neighborhood_description": row["neighborhood_description"],
        "description_vector": get_embedding(row["description"]),
    },
    axis=1,
).tolist()

# Create or load the table
table_name = "listings"
if table_name in db.table_names():
    table = db.open_table(table_name)
else:
    table = db.create_table(table_name, schema=RealEstateListings)

try:
    table.add(data)
except Exception as e:
    print("Error inserting data:", e)

# View a sample of the table
df = table.head(table.count_rows()).to_pandas()
df.to_csv("listings_csv.csv")
df.head(20)
```

## 4. Collecting Buyer Preferences

```python
# Buyer preferences
preferences_1 = [
    "A comfortable three-bedroom house with a spacious kitchen.",
    "Quiet neighborhood, good schools, close to shopping.",
    "A backyard, energy-efficient systems, garage.",
    "Suburban area with urban amenities nearby.",
    "$750,000"
]

preferences_2 = [
    "A cozy 1-bedroom apartment with a modern kitchen and city views.",
    "Close to public transport, restaurants, and entertainment.",
    "Pet-friendly with a small balcony for outdoor space.",
    "Urban neighborhood with a vibrant community and easy access to downtown.",
    "$350,000"
]
```

## 5. Semantic Search for Real Estate Listings Based on Buyer Preferences

```python
class RealEstateListing(LanceModel):
    location: str
    list_price: float
    bedrooms: float
    bathrooms: float
    square_feet: float
    description: str
    neighborhood_description: str
    description_vector: list

def get_top_recommendation(preferences: list) -> tuple:
    # Convert the buyer's preferences into a list of embeddings
    preference_embeddings = [embeddings.embed_query(pref) for pref in preferences]

    # Aggregate embeddings by summing them element-wise using numpy
    final_preference_embedding = np.sum(preference_embeddings, axis=0)

    # Use the table.search method to retrieve the most similar listings (limit to 1 for top match)
    results = table.search(final_preference_embedding).limit(1).to_pydantic(RealEstateListing)

    # Extract the most relevant listing (top 1)
    top_listing = results[0]
    
    # Return the relevant information (location, price, description)
    return top_listing.location, top_listing.list_price, top_listing.description
```

## 6. Augmented Response Generation

```python
# Augment the top listing based on buyer preferences
def augment_listing_description(location: str, price: float, description: str, preferences: list) -> str:
    # Customize the prompt based on buyer preferences
    augmented_prompt = f"""
    Generate a compelling, personalized description for a property located in {location}, priced at ${price}.
    The buyer values {preferences[0]} (e.g., spacious kitchens), prefers a neighborhood with {preferences[1]} (e.g., quiet, family-friendly), and is looking for homes with {preferences[2]} (e.g., backyard pool).
    Highlight the amenities and lifestyle features that align with these preferences, making it an irresistible option for this buyer.
    """

    # Generate an augmented description using the LLM
    augmented_description = llm.predict(augmented_prompt)
    return augmented_description
```

## 7. Demo

```python
# Example of augmenting the top recommendation based on preferences_1
top_listing = get_top_recommendation(preferences_1)
location, price, description = top_listing

# Augment the top listing's description with buyer preferences
augmented_description = augment_listing_description(location, price, description, preferences_1)

# Print the top listing and its augmented description
print("Top Listing: ", (location, price, description))
print("Augmented Description: ", augmented_description)
```
